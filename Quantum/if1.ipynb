```python
from qiskit_aer import AerSimulator
from qiskit import QuantumCircuit, QuantumRegister,ClassicalRegister, transpile

from IPython.display import Math
from qc_workbook.show_state import statevector_expr

import numpy as np

print("----Start----")
simulator = AerSimulator(method='statevector')
print(simulator.name)
```

```python
register1 = QuantumRegister(4, name='reg1')
register2 = QuantumRegister(1, name='reg2')
output = ClassicalRegister(4, name='out1')

circuit = QuantumCircuit(register1, register2, output)
circuit.h(register1)
```

```python
circuit.measure(register1, output)
```

```python
dtheta = 2. * np.pi / 16.

for idx in range(4):
    angle = dtheta*(2**idx)
    with circuit.if_test((output[idx], 1)):
        circuit.ry(angle, register2[0])

circuit.draw('mpl')
```

```python
Math(statevector_expr(circuit, register_sizes=[4, 1]))
```
